# PYTHON PRINT STATEMENTS

Printing in python is really simple we output using 'print' function.

```python
# use print function
print('Python')
```

In the above line, the value inside the quotation is a string, <br>

in python strings are always written inside either single, double or triple quotation marks.

```python
# double quotation marks example
print("python")
# triple quotation marks exmaple
print('''Python''')
```

```python
# wrong example with different quotation marks
print("python")
```

In the above case we just print single word strings though we can also print a complete line as well. The line is also an example of string.

```python
# print a line
print("Python is awesome")
```

```python
print('''Python is awesome
just a demo.''')
```

Other than printing strings we can also print numbers.<br>
we have majorly two types of numbers in python:<br>
* integers - 1, 3, -99, 100
* floats or decimal numbers - 9.9, -77.8, 0.1, 1.0

let's print them here

```python
# print an integer
a='8'
print(a)
print(type(a))

# print a float
print(0.2)
```
